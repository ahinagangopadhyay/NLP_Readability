```python
import spacy

import textstat  # for easier access to top-level functions

# Load spaCy model once
nlp = spacy.load('en_core_web_sm')

# Sentence splitting using spaCy
def break_sentences(text):
    doc = nlp(text)
    return list(doc.sents)

# Count words
def word_count(text):
    sentences = break_sentences(text)
    return sum(len([token for token in sentence]) for sentence in sentences)

# Count sentences
def sentence_count(text):
    return len(break_sentences(text))

# Average sentence length
def avg_sentence_length(text):
    words = word_count(text)
    sentences = sentence_count(text)
    return float(words / sentences) if sentences > 0 else 0

# Syllable count of a word
def syllables_count(word):
    return textstatistics().syllable_count(word)

# Average syllables per word
def avg_syllables_per_word(text):
    syllable = textstatistics().syllable_count(text)
    words = word_count(text)
    return round(float(syllable) / float(words), 1) if words > 0 else 0

# Count difficult words
def difficult_words(text):
    words = [token.text for token in nlp(text) if not token.is_stop]
    return len([word for word in words if syllables_count(word) >= 2])

# Count polysyllabic words
def poly_syllable_count(text):
    words = [token.text for token in nlp(text)]
    return sum(1 for word in words if syllables_count(word) >= 3)

# Flesch Reading Ease
def flesch_reading_ease(text):
    FRE = 206.835 - (1.015 * avg_sentence_length(text)) - \
          (84.6 * avg_syllables_per_word(text))
    return round(FRE, 2)

# Gunning Fog Index
def gunning_fog(text):
    per_diff_words = (difficult_words(text) / word_count(text) * 100) if word_count(text) > 0 else 0
    grade = 0.4 * (avg_sentence_length(text) + per_diff_words)
    return round(grade, 2)

# SMOG Index
def smog_index(text):
    if sentence_count(text) >= 3:
        poly_syllab = poly_syllable_count(text)
        SMOG = (1.043 * ((30 * (poly_syllab / sentence_count(text))) ** 0.5)) + 3.1291
        return round(SMOG, 1)
    return 0

# Dale-Chall Score
def dale_chall_readability_score(text):
    words = word_count(text)
    difficult = difficult_words(text)
    if words == 0:
        return 0
    percent_difficult = (difficult / words) * 100
    raw_score = (0.1579 * percent_difficult) + (0.0496 * avg_sentence_length(text))
    if percent_difficult > 5:
        raw_score += 3.6365
    return round(raw_score, 2)

# Run all stats and print them
def main():
    # Example sample text – replace with file reading if needed
    sample_text = """
    The sun is bright. The sky is blue. Birds fly in the air. 
Children play in the park. Trees are green. Flowers bloom in spring. 
Water is wet. Ice is cold. Dogs bark. Cats sleep. 
We eat food to live. We drink water to stay healthy. 
Every day is a new chance to smile.

    """

    print("Flesch Reading Ease:", flesch_reading_ease(sample_text))
    print("SMOG Index:", smog_index(sample_text))
    print("Gunning Fog Index:", gunning_fog(sample_text))
    print("Dale-Chall Score:", dale_chall_readability_score(sample_text))

if __name__ == "__main__":
    main()
```
